```python
# Importar archivo
import pandas as pd

df = pd.read_csv('../data/shopping_trends_updated.csv')

print(f'Dimensiones del dataframe original {df.shape}')
df.head()
```

Se crean nuevas variables `Total gastado` y `Cliente VIP` que pueden ser de utilidad para proponer una segmentación de clientes al área de Marketing. 

Por ejemplo: Una campaña de Fidelización con desceuntos exclusivos a los clientes que más gastan e incentivar a que continuen comprando.

```python
# Total gastado: Se agrupan los Clientes por ID para sumar todas las compras que han hecho en esta nueva columna
df['Total gastado'] = df.groupby('Customer ID')['Purchase Amount (USD)'].transform('sum')

# Cliente VIP: Se compara la nueva columna 'Total gastado' para crear esta nueva columna booleana que indicará True si el cliente es parte del 50% que más gasta
df['Cliente VIP'] = df['Total gastado'] > df['Total gastado'].median()

print(f'Dimensiones del dataframe modificado {df.shape}')
df.sort_values(by='Total gastado', ascending=False).head()
```

```python
# Se cuenta el número de clientes VIP (Cliente VIP == True)
cantidad_clientes_vip = df['Cliente VIP'].sum()

# Se muestra el resultado
print("Cantidad de Clientes VIP (True):", cantidad_clientes_vip)

# ¿Qué elementos son los que más compran estos clientes?


clientes_vip = df[df['Cliente VIP'] == True]

# Contar las ocurrencias de cada producto
mas_comprado = clientes_vip['Item Purchased'].value_counts()

# Mostrar los productos más comprados
print("Productos más comprados por los clientes VIP:")
print(mas_comprado)
```

Como campaña se propone ofrecer desceuntos a los clientes VIP, en los productos más comprados, de acuerdo a la temporada (invierno o verano) en curso.  
Los principales 5 elementos que los clientes VIP llevan son: 
- Shirt         91
- Dress         88
- Blouse        88
- T-shirt       84
- Pants         83

Otra campaña podría considerar la información de la columna `Previous Purchases` que indica la cantidad de compras que ha tenido un cliente en la historia.

Para lo anterior, se crea la variable booleana `Cliente frecuente` que nos indica si el cliente corresponde al 50% que más veces ha compado

```python
df['Cliente frecuente'] = df['Previous Purchases'] > df['Previous Purchases'].median()

# Se cuenta el número de clientes frecuentes (Cliente frecuente == True)
cantidad_clientes_frecuente = df['Cliente frecuente'].sum()

# Se muestra el resultado
print(f'Dimensiones del dataframe modificado 2 {df.shape}')
print("Cantidad de Clientes frecuentes (True):", cantidad_clientes_frecuente)
```
