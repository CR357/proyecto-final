```python
# Importar archivo
import pandas as pd

df = pd.read_csv('../data/shopping_trends_updated.csv')
```

### Análisis de correlaciones  
Se analizará correlaciones entre las variables, para identificar patrones significativos y variables de interés en el dataframe

```python
# Cual es la correlación entre la edad y la cantidad de ultimas compras?

correlation_edad_compras_anteriores = df['Age'].corr(df['Previous Purchases'], method='pearson')
print("Coeficiente de Correlación de Pearson:", correlation_edad_compras_anteriores)# No se ve correlación significativa
```

### Matriz de correlación y scatterplots

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

## Por facilidad, en primer lugar se trabaja con las columnas numéricas
columnas_selec = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
columnas_corr = df[columnas_selec]

matriz_corr = columnas_corr.corr()

# Se  crea la matriz de correlaciones de las variables seleccionadas usando matplotlib y seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlaciones')
plt.show()

# se crea un scatter plot con seaborn a las variables con mayor correlación 
# corr = 0.031 para 'Review Rating' y 'Purchase Amount (USD)'
# corr = 0.04 para 'Age' y 'Purchase Amount (USD)'
sns.scatterplot(data=df, x='Review Rating', y='Purchase Amount (USD)')
plt.title('Review Rating vs Purchase Amount (USD)')
plt.xlabel('Review Rating')
plt.ylabel('Purchase Amount (USD)')
plt.show()

sns.scatterplot(data=df, x='Age', y='Purchase Amount (USD)')
plt.title('Age vs Purchase Amount (USD)')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```
